```python
%%manim --format=gif --save_last_frame -r 1920,216 -qh AllLeadsShuffled
# 
class AllLeadsShuffled(Scene):
    
    def construct(self):
        image_width = 1920/135
        image_height = 216/135
        dist_between_rect = 0.2
        total_rect = int(20)
        rect_width = (image_width - (total_rect + 1) * dist_between_rect) / total_rect
        
        x_position = 1/2*(1 - total_rect)*(rect_width + dist_between_rect)
        
        fraud_rectangles = []
        for i in range(total_rect):
            rectangle = Rectangle(
                color="#FFFFFF",
                height = 1.3,
                width = rect_width
            )
            rectangle.set_fill(WHITE, opacity=1.0)
            rect_obj = {"geom": rectangle, "value": None}
            rectangle.move_to(np.array([x_position, 0, 0]))
            fraud_rectangles.append(rect_obj)
            x_position = x_position + dist_between_rect + rect_width
        
        for i, rectangle in enumerate(fraud_rectangles):
            r = rectangle["geom"]
            self.add(r)
            
            if i in [4,12,14,18]:
                image = ImageMobject("./img/skull.png")
                image.height = 0.5
                # Change to align
                image.align_to(r, LEFT)
                self.add(image)
        self.wait()
        # circle.set_fill(mn.WHITE, opacity=1.0)  # set color and transparency

        # square = Square()  # create a square
        # square.rotate(mn.PI / 4)  # rotate a certain amount
        

        # self.play(Create(square))  # animate the creation of the square
        # self.play(Transform(square, circle))  # interpolate the square into the circle
        # self.play(FadeOut(square))  # fade out animation
```
