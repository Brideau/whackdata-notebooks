```python
%%manim --format=gif --save_last_frame -r 680,514 -qh MainTitle
# 
class MainTitle(Scene):
    
    def construct(self):
        image_width = 1920/135
        image_height = 216/135
        dist_between_rect = 0.2
        total_rect = int(20)
        rect_width = (image_width - (total_rect + 1) * dist_between_rect) / total_rect
        
        x_position = 1/2*(1 - total_rect)*(rect_width + dist_between_rect)
        
        fraud_rectangles = []
        for i in range(total_rect):
            rectangle = Rectangle(
                color="#FFFFFF",
                height = 2.0,
                width = rect_width
            )
            rectangle.set_fill(WHITE, opacity=1.0)
            rect_obj = {"geom": rectangle, "value": None}
            rectangle.move_to(np.array([x_position, 0, 0]))
            fraud_rectangles.append(rect_obj)
            x_position = x_position + dist_between_rect + rect_width
        
        for i, rectangle in enumerate(fraud_rectangles):
            r = rectangle["geom"]
            self.add(r)
        
        font_size = 500
        font="Charter"
        title_text = Text("P@k", font_size=font_size, color="#262626", font=font)
        
        self.add(title_text)
        self.wait()
```

```python
%%manim --format=gif --save_last_frame -r 1920,324 -qh LargeLeadBatch
# 
class LargeLeadBatch(Scene):
    
    def construct(self):
        dist_between_rect = 0.1
        total_rect = int(100)
        rect_width = (image_width - (total_rect + 1) * dist_between_rect) / total_rect
        
        x_position = 1/2*(1 - total_rect)*(rect_width + dist_between_rect)
        
        fraud_rectangles = []
        
        brace_1_pos = []
        brace_2_pos = []
        
        for i in range(total_rect):
            rectangle = Rectangle(
                color="#FFFFFF",
                height = 0.8,
                width = rect_width
            )
            if i < 20:
                color = BLACK
            else:
                color = WHITE
            rectangle.set_fill(color, opacity=1.0)
            rectangle.set_stroke(color=color, width=1)
            rectangle.move_to(np.array([x_position, 0.3, 0]))
            fraud_rectangles.append(rectangle)
            
            if i == 0:
                brace_1_pos.append(rectangle.get_top())
            elif i == 19:
                brace_1_pos.append(rectangle.get_top())
            elif i == 20:
                brace_2_pos.append(rectangle.get_top())
            elif i == 99:
                brace_2_pos.append(rectangle.get_top())
            x_position = x_position + dist_between_rect + rect_width
        
        for i, rectangle in enumerate(fraud_rectangles):  
            self.add(rectangle)
        
        brace_shift = -0.7
        brace_1 = BraceBetweenPoints(
            brace_1_pos[0] + np.array([0, brace_shift, 0]),
            brace_1_pos[1] + np.array([0, brace_shift, 0])
        )
        brace_2 = BraceBetweenPoints(
            brace_2_pos[0] + np.array([0, brace_shift, 0]),
            brace_2_pos[1] + np.array([0, brace_shift, 0])
        )
        self.add(brace_1, brace_2)
        
        font_size = 27
        font="Charter"
        left_text = Text("Leads Processed", font_size=font_size, color=WHITE, font=font)
        left_text.move_to(np.array([-5.65, -0.8, 0]))
        
        right_text = Text("Leads Above Capacity", font_size=font_size, color=WHITE, font=font)
        right_text.move_to(np.array([1.5, -0.8, 0]))
        
        self.add(left_text, right_text)
        
        self.wait()
```

```python
%%manim --format=gif --save_last_frame -r 1920,290 -qh GoodVBad

from numpy.random import default_rng
rng = default_rng()

class GoodVBad(Scene):
    
    def construct(self):
        dist_between_rect = 0.05
        total_rect = int(150)
        rect_width = (image_width - (total_rect + 1) * dist_between_rect) / total_rect
        
        x_position = 1/2*(1 - total_rect)*(rect_width + dist_between_rect)
        
        fraud_rectangles = []
        
        brace_1_pos = []
        brace_2_pos = []
        
        for i in range(total_rect):
            rectangle = Rectangle(
                color="#FFFFFF",
                height = 0.8,
                width = rect_width
            )
            is_good = rng.binomial(1, (total_rect - i) / total_rect)
            if is_good:
                color = "#2e5497"
            else:
                color = WHITE
            rectangle.set_fill(color, opacity=1.0)
            rectangle.set_stroke(color=color, width=1)
            rectangle.move_to(np.array([x_position, 0.3, 0]))
            fraud_rectangles.append(rectangle)
            
            if i == 0:
                brace_1_pos.append(rectangle.get_top())
            elif i == 49:
                brace_1_pos.append(rectangle.get_top())
            x_position = x_position + dist_between_rect + rect_width
        
        for i, rectangle in enumerate(fraud_rectangles):  
            self.add(rectangle)
        
        brace_shift = -0.7
        brace_1 = BraceBetweenPoints(
            brace_1_pos[0] + np.array([0, brace_shift, 0]),
            brace_1_pos[1] + np.array([0, brace_shift, 0])
        )
        self.add(brace_1)
        
        font_size = 27
        font="Charter"
        left_text = Text("k = 40", font_size=font_size, color=WHITE, font=font, slant=ITALIC)
        left_text.move_to(np.array([-4.7, -0.7, 0]))
        
        
        self.add(left_text)
        
        self.wait()
```
